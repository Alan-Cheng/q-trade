```python
import itertools
```

```python
items = [1,2,3]
# permutiations： 生成所有組合
for item in itertools.permutations(items):
    print(item)

# combinations()： Cn取x
for item in itertools.combinations(items, 2):
    print(item)

# product()： 生成笛卡爾積
ab = ['a', 'b']
cd = ['c', 'd']
for item in itertools.product(ab, cd):
    print(item)
```
